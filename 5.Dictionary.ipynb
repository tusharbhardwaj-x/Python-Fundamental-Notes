# Dictionary{}:
### It is a data type that stores multiple values in a single variable but in form of pair.
### Mutable: Values present in dictionary are changeable.
### Ordered data: Output will same as the input.
### Pair: {Keys:Values}

### access: To access all keys, all values, specific key, specific value, add, update, etc.

```python
Smartphone = {"Name":"Note30", "Brand":"Infinix", "Price":15999, "RAM":"8GB"}

# access all keys
a = Smartphone.keys()
print(a)
print(Smartphone.keys())

# access all values
a = Smartphone.values()
print(a)
print(Smartphone.values())

# access specific value with key
a = Smartphone.get("Name")
print(a)
print(Smartphone["Name"])
a = Smartphone["Name"]
print(a)

# access complete set
print(Smartphone)
a = Smartphone.items()
print(a)
print(Smartphone.items())

# update w/o using method
Smartphone["Price"] = 14999
print(Smartphone)

# add w/o using method
Smartphone["Storage"] = "256GB"
print(Smartphone)
```

### copy: To copy dictionary from one variable to other

```python
Smartphone = {"Name":"Note30", "Brand":"Infinix", "Price":15999, "RAM":"8GB"}
mobile = Smartphone.copy()
print(mobile)
print(Smartphone)
```

### update: To update any value in a dictionary

```python
mob = {"Name":"Note30", "Brand":"Infinix", "Price":15999, "RAM":"8GB"}
mob.update({"RAM":"12GB"})
print(mob)
```

### pop: To remove any pair by their key

```python
mob = {"Name":"Note30", "Brand":"Infinix", "Price":15999, "RAM":"8GB"}
mob.pop("RAM")
print(mob)
```

### clear: To remove all items from the dictionary

```python
mob = {"Name":"Note30", "Brand":"Infinix", "Price":15999, "RAM":"8GB"}
mob.clear()
print(mob)
```

### del: To remove all itmes from the list as well as structer of the dictionary

```python
mob = {"Name":"Note30", "Brand":"Infinix", "Price":15999, "RAM":"8GB"}
del mob()
print(mob)
```

### loop: To ittrate the keys, values, items of the dictionary

```python
mob = {"Name":"Note30", "Brand":"Infinix", "Price":15999, "RAM":"8GB"}
for x in mob:
    print(x)
for x,y in mob.items():
    print(x)
    print(y)
    print(x, y)
```

## nested dictionary: Dictionaries inside dictionary

```python
shop = {"Phone1":{"Name":"Note30", "Brand":"Infinix", "Price":15999, "RAM":"8GB"},
        "Phone2":{"Name":"G85", "Brand":"Motorolla", "Price":12999, "RAM":"4GB"},
        "Phone3":{"Name":"M5", "Brand":"Poco", "Price":13999, "RAM":"4GB"},
        "Phone4":{"Name":"Narzo50", "Brand":"Realme", "Price":16999, "RAM":"8GB"},
        "Phone5":{"Name":"16Pro", "Brand":"Iphone", "Price":119999, "RAM":"8GB"}}
print(shop.items())
print(shop)

for x,y in shop["Phone1"].items():
    print(x,y)
for x,y in shop["Phone5"].items():
    print(x,y)

print(shop["Phone1"])
```
