# List []: 
## A data type used to store multiple item in a single variable in python
## Mutable : Content are changeable
## ordered : Same output as we inputted
## duplicate values are allowed

### insert : to insert any single value in a list within a specific indexing

```python
a = ["Tushar", "Rahul", "Himanshu"]
a.insert(1,"Kartik")
print(a)
```

```python
a = ["Buggati", "Bullet", "Duke", 95, 43, "Swift"]
a.insert(2,56)
print(a)
```

### append: To insert any single value in a list at end position

```python
a = ["Neha", "Pooja", "Aditiya", "Rahul"]
a.append("Badal")
print(a)
```

```python
a = ["Noni", 43, 53, 43, 34, "Om"]
a.append(4)
print(a)
```

### extend: To join two lists together

```python
a = [32, 43, 45, 43, 54, 76, 74]
b = [45, 65, 56, 65, 65]
a.extend(b)
print(a)
```

```python
a = [32, 43, 45, 43, 54, 76, 74]
b = ["T", "U", "S", "H", "A", "R"]
a.extend(b)
print(a)
```

### remove: To direct remove any value from the list

```python
a = [32, 43, 45, 43, 54, 76, 74]
a.remove(43)
print(a)
```

```python
a = ["Adi", "Raj", "Neha", "Ram", "Rohan"]
a.remove("Rohan")
print(a)
```

### pop: To remove any value from the list by their indexing

```python
a = ["Om", 43, "Tushar", "Kartik", "Lovekesh"]
a.pop(1)
print(a)
```

### delete: To remove multiple value from the list by their indexing

```python
a = ["Om", 43, "Tushar", "Kartik", "Lovekesh"]
del a[1:4]
print(a)
```

```python
a = [23, 596, 9, 6, 89 , 96, 59, 596, 485]
del a[1:]
print(a)
```

### clear: To remove all values of lsit

```python
a = [34, 54, 564, 35, 45]
a.clear()
print(a)
```

```python
a = ["B", "H", "A", "R", "D", "W", "A", "J"]
a.clear()
print(a)
```

### sort: To arrange the list in asscending order

```python
a = ["Om", "Tushar", "Kartik", "Lovekesh"]
a.sort()
print(a)
```

```python
a = [23, 596, 9, 6, 89 , 96, 59, 596, 485]
a.sort()
print(a)
```

### sort reverse: To arrage list in descending order

```python
a = ["Om", "Tushar", "Kartik", "Lovekesh"]
a.sort(reverse = True)
print(a)
```

```python
a = [23, 596, 9, 6, 89 , 96, 59, 596, 485]
a.sort(reverse= True)
print(a)
```

### copy: To copy list from one variable to another

```python
a = ["Om", "Tushar", "Kartik", "Lovekesh"]
b = a.copy()
print(b)
```

```python
a = [72, 89, 97, 56, 26, 59]
b = a.copy()
print(a)
print(b)
```

### count: To count no. of repetation of a value in a list

```python
a = [32, 43, 45, 43, 54, 76, 74]
b = a.count(43)
print(b)
```

```python
a = ["B", "H", "A", "R", "D", "W", "A", "J"]
b = a.count("A")
print(b)
```

### index: To find indexing of any value in a list

```python
a = [32, 46, 45, 43, 54, 76, 74]
b = a.index(43)
print(b)
```

```python
a = ["B", "H", "A", "R", "D", "W", "A", "J"]
b = a.index("B")
print(b)
```

### len: To find total no. of values present in a list

```python
a = [32, 46, 45, 43, 54, 76, 74]
b = len(a)
print(b)
```

```python
a = ["B", "H", "A", "R", "D", "W", "A", "J"]
b = len(a)
print(b)
```

### type: To check the data type

```python
a = ["B", "H", "A", "R", "D", "W", "A", "J"]
b = type(a)
print(b)
```
