# Logical Operator: To use logics along with the multiple variables and always give output in boolean data type

### 1.and: To check both conditions are satisfied or not

```python
a = 34
b = 43
c = 34
d = a==c and b>c
print(d)
print(a>=c and b==a)
print(a==b and b==c)
```

### 2.or: To check either of the condition is satisfied or not

```python
a = 34
b = 43
c = 34
d = a==c or b>c
print(d)
print(a>=c or b==a)
print(a==b or b==c)
```

### 3.not: To oppose both the condition we studied above

```python
a = 34
b = 43
c = 34
d = not(a==c and b>c)
print(d)
print(not(a>=c and b==a))
print(not(a==b and b==c))
```

```python
a = 34
b = 43
c = 34
d = not(a==c or b>c)
print(d)
print(not(a>=c or b==a))
print(not(a==b or b==c))
```

# Identity Operator: To compare the values as well as location of the multiple  varibles and give output in boolean data type

### 1.is: To check both location and values of the variable are same or not

```python
a = [34,92,33]
b = [34,92,33]
c = a
print(a is b)
print(c is a)
```

### 2.is not: Just opposite of is

```python
a = [34,92,33]
b = [34,92,33]
c = a
print(a is not b)
print(c is not a)
```

# Membership Operators: To check the given value is the part of the variable or not and give output in boolean data type

### 1.in: To check the value present in variable or not

```python
a = [34,92,33]
b = 34
c = 67
print(b in a)
print(c in a)
```

### 2.not in: Just opposite of in

```python
a = [34,92,33]
b = 34
c = 67
print(b not in a)
print(c not in a)
```

# Bitwise Operators: To find any number in bits i.e, totaly work in computer language i.e, only understand 0(off) and 1(on), Calculation performed OTBO Truth table

### 1.and: & if both on then on otherwise off [basically first convert variables into their bitwise form then multiplication performed]

```python
a = 5
b = 3
print(a&b)
```

### 2.or: | if either of the variable is on then on otherwise off [basically first convert variables into their bitwise form then addition performed]

```python
a = 5
b = 3
print(a|b)
```

### 3.xor: ^ if either of the variable is different then on otherwise off

```python
a = 5
b = 3
print(a^b)
```

### 4.not: ~ -[n+1]

```python
a = 5
print(~(a))
```

### 5.right shift: >> shift the bitwise format right side and add off values in left side according to how much values you shift

```python
a = 5
b = 3
print(a>>b)
```

### 6.left shift: << shift the bitwise format left side and add off values in right side according to how much values you shift

```python
a = 5
b = 3
print(a<<b)
```
