# For Loop: To ittrate the values present in data types

```python
a = "Tushar"
for x in a:
    print(x)

b = ["list", "tuple", "set", "dictionary"]
for x in b:
    print(x)
```

### end: ittrate but in row

```python
a = ["Under", "Line", "the", "topics", 345, 34]
for x in a:
    print(x, end=" ")
```

### break: To break the ittration when the conditon is satisfied

```python
car = ["Mahindra", "Tata", "Tesla", "Ferrari", "Hyndia", "Maruti", "Wolswagon"]
for x in car:
    print(x)
    if x=="Ferrari":
        break
```

```python
car = ["Mahindra", "Tata", "Tesla", "Ferrari", "Hyndia", "Maruti", "Wolswagon"]
for x in car:
    if x=="Ferrari":
        break
    print(x)
```

### continue: To skip ittration for a specific value as per condition

```python
bike = ["Bajaj", "Hero", "Honda", "Ktm", "Royal Enfield", "Harley Davidson", "Yamaha"]
for x in bike:
    if x=="Ktm":
        continue
    print(x)
```

```python
car = ["Mahindra", "Tata", "Tesla", "Ferrari", "Hyndia", "Maruti", "Wolswagon"]
for x in car:
    if x=="Hyndia":
        continue
    print(x)
```

### range: use to ittrate sequencial value or value with specific pattern as per our condition

```python
for x in range(10):
    print(x)
```

```python
for x in range(2,10):
    print(x)
```

```python
for x in range(7,71,7):
    print(x)
```

```python
num = int(input("Table of: "))
for x in range(1,11):
    print(num,"x", x, "=", num*x)
```

## patterns:

```python
rows = 5
for a in range(1,rows+1):
    for b in range(a):
        print("*", end=" ")
    print()
```

```python
rows = 5
for a in range(rows, 0, -1):
    for b in range(a):
        print("*", end=" ")
    print()
```

```python
rows=5
for a in range(1, rows+1):
    for b in range(rows-a):
        print(" ", end=" ")
    for c in range(a):
        print(" * ", end=" ")
    print()
```

```python
rows = 5
for a in range(rows, 0, -1):
    for b in range(rows-a):
        print(" ", end="")
    for c in range(a):
        print("* ", end=" ")
    print()
```

# While Loop: Use to ittrate till a specific condition satisfied

```python
a = 1
while a<10:
    print(a)
    a+=1
```

```python
a = 1
while a<10:
    a+=1
    print(a)
```

```python
a = 0
while a<101:
    if a%2==0:
        print(a)
    a+=1
```

### break: To stop ittration when a specific condition satisfied

```python
a = 1
while a<25:
    print(a)
    if a==15:
        break
    a+=1
```

### continue: To skip ittration at some value as per specific condition

```python
a = 11
while a<20:
    a+=1
    if a==15:
        continue
    print(a)
```
