# Math: Python has inbuilt module math which provide various mathematical functions.

```python
import math
```

```python
# constant: Some predefined values in maths like pi, e, tau, nan etc.
a = math.pi
print(a)
b = math.e
print(b)
print(math.tau)
print(math.nan)
```

```python
# ceil & floor: To get the value more than decimal and less then decimal respectively.
a = math.ceil(43.93)
print(a)
b = math.floor(43.93)
print(b)
```

```python
# airthmetic functions: prod, pow, sqrt, log etc.
a= [43,65,43,56]
print(math.prod(a))
print(math.pow(13,2))
print(math.sqrt(169))
print(math.log(1))
```

```python
# trignometric functions: sin, cos, tan, radians etc.
print(math.sin(90))
print(math.cos(90))
print(math.tan(60))
print(math.radians(180))
print(math.fabs(-473))    # mod
```

# Date & Time: It provide access to various date, time, day, hours, min, sec, formats.

```python
import datetime
```

```python
# Basics: current date and time, year, month, day, hours, etc.
a = datetime.datetime.now()
print(a)
print(a.year)
print(a.month)
print(a.day)
print(a.hour)
print(a.minute)
print(a.second)
```

```python
# small alphabets: except (i,k,l,n,o,q,t,v,z), all small alphabets has their predefined formats
a = datetime.datetime.now()
print(a)
print(a.strftime("%a"))
print(a.strftime("%b"))
print(a.strftime("%c"))
print(a.strftime("%d"))
print(a.strftime("%e"))
print(a.strftime("%f"))
print(a.strftime("%g"))
print(a.strftime("%h"))
print(a.strftime("%j"))
print(a.strftime("%m"))
print(a.strftime("%p"))
print(a.strftime("%r"))
print(a.strftime("%u"))
print(a.strftime("%w"))
print(a.strftime("%x"))
print(a.strftime("%z"))
```

```python
# capital alphabets: except (E,J,K,L,N,O,P,Q,Z), all Capital alphabets has their predefined formats
a = datetime.datetime.now()
print(a)
print(a.strftime("%A"))
print(a.strftime("%B"))
print(a.strftime("%C"))
print(a.strftime("%D"))
print(a.strftime("%F"))
print(a.strftime("%G"))
print(a.strftime("%H"))
print(a.strftime("%I"))
print(a.strftime("%M"))
print(a.strftime("%R"))
print(a.strftime("%S"))
print(a.strftime("%T"))
print(a.strftime("%U"))
print(a.strftime("%V"))
print(a.strftime("%W"))
print(a.strftime("%X"))
print(a.strftime("%Y"))
```

# Random: It is used to generate pseudo numbers or random choises.

### randint: To get random numbers according to the given range.

```python
import random
```

```python
a = random.randint(1,6)
print(a)
b = random.randint(5,10)
print(b)
```

### choice: To get random variable according to given choices.

```python
a = random.choice(["Head", "Tail"])
print(a)
b = random.choice(["Tushar", "Himanshu", "Kartik", "Rahul"])
print(b)
```
