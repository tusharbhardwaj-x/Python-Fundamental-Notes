# if: It always holds true conditions

```python
marks = 93
if marks>90:
    print("pass")
if marks<90:
    print("fail")
```

### user input in if

```python
marks = int(input("Enter Your Marks: "))
if marks>=33:
    print("Pass")
```

# if else: It always holds two conditions.

```python
marks = 53
if marks>50:
    print("Good")
else:
    print("Bad")

if marks>70:
    print("Excellent")
else:
    print("Better luck Next time")
```

### user input in if-else

```python
marks = int(input("Enter Your Marks: "))
if marks>=60:
    print("Good")
else:
    print("Bad")
```

# elif: It holds multiple conditions

```python
marks = 45
if marks>=90:
    print("A")
elif marks>=70:
    print("B")
elif marks>=50:
    print("C")
elif marks>=33:
    print("D")
else:
    print("F")
```

### user input in elif

```python
marks = int(input("Enter Your Marks: "))
if marks>=90:
    print("A")
elif marks>=70:
    print("B")
elif marks>=50:
    print("C")
elif marks>=33:
    print("D")
else:
    print("F")
```

# Nested-if: conditions inside condition

### login page

```python
usernames = ["toshii.x", "uc.anshu", "kartik.xu","amanxu_01", "tanuu.ii", "payalll_1_", "cahvimishra13"]
passwords = ["A7k3L9p2Q1", "m4R8zW2vY6", "B2a9K7PqL5", "X3y8M2kR7z","p7Q2L4vT9s","Z8xK2mP6Q4","t9R2vW7aL3"]
data = dict(zip(usernames, passwords))

username = str(input("Username"))
if username in data:
    password = str(input("Password"))
    if password == data[username]:
        print(username, "successfully logged in")
    else:
        print("Incorrect Password try again")
else:
    print("No user found")
```

### odd-even check

```python
num = int(input("Enter the number: "))
if num>0:
    print("Positive")
    if num%2==0:
        print("Even")
    else:
        print("Odd")
else:
    print("Negative")
```

# Short hand if: To use conditional statement in single line or shortly.

```python
age = int(input("Enter Your Age: "))
print("Eligible") if age>=18 else ("Not Eligible")
```
