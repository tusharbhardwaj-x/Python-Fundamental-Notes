# Numeric Type: Stores numeric data

### int: To store whole numbers

```python
a = 56
b = 8493
print(type(a))
print(type(b))
```

### float: To store decimal numbers

```python
a = 55.35
b = 840.00
print(a)
print(b)
print(type(a))
print(type(b))
```

### complex: To store imaginary numbers

```python
a = 45+3j
b = 8+93j
print(a)
print(b)
print(type(a))
print(type(b))
```

# Sequence Type: Stores multiple values in orders

### string"": To store words or sentences

```python
a = "Tushar"
b = "I live in Rohini"
print(a)
print(b)
print(type(a))
print(type(b))
```

### List[]: To store multiple value in single variable

```python
a = ["Tushar", 56, 34.90, 23+45j]
b = ["T", 543, 543, 765, 32]
print(a)
print(b)
print(type(a))
print(type(b))
```

### tuple(): To store multiple value in single variable

```python
a = ("Tushar", 56, 34.90, 23+45j)
b = ("T", 543, 543, 765, 32)
print(a)
print(b)
print(type(a))
print(type(b))
```

# Set Type: Stores multiple values but unordered and w/o duplicates

### set{}: To store multiple value in single variable

```python
a = {"Tushar", 56, 34.90, 23+45j}
b = {"T", 543, 564, 765, 32}
print(a)
print(b)
print(type(a))
print(type(b))
```

### frozenset([]): immutable version of set

```python
a = frozenset(["Tushar", 56, 34.90, 23+45j])
b = frozenset(["T", 543, 564, 765, 32])
print(a)
print(b)
print(type(a))
print(type(b))
```

# Mapping Type: Store value in the format of keys and values i.e, pairs

### Dictionary{}: To store multiple pairs in single variable

```python
a = {"str":"Tushar", "int":56, "float":34.90, "complex":23+45j}
b = {"a":"T", "b":543, "c":564, "d":765}
print(a)
print(b)
print(type(a))
print(type(b))
```

# Boolean Type: Check conditions and output will TRUE or FALSE

```python
a = True
b = False
print(a)
print(b)
print(type(a))
print(type(b))
```
