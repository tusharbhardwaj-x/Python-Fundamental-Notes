# function:
### A block of code & when we call fucntion we got result.
### Parameters: Given during cretion of a function, inside paranthesis.
### Arguments: Given during call inside parenthesis leads to inverted commas.

```python
def Chief():            # Chief: Defined as function
    print("Tushar")
Chief()
```

```python
Chief()
```

```python
def add(a,b):        # a,b: Parameters
    print(a+b)
add(34,67)
```

```python
add(849,854)
```

```python
def Intro(name, age):
    print("My name is", name, "I'm", age, "years old.")
Intro("Tushar", 20)             # Tushar, 20: Arguments
```

```python
Intro("Himanshu", 18)
```

### arbitary functions: when we want to call multiple arguuments in a single funnction

```python
def Wish(*names):
    print("Good Morning", names)
    print("Good Morning", names[2])
    print("Good Morning", names[1])
Wish("Tushar", "Kartik", "Himanshu", "Rahul")
```

```python
def intro(*name,age):
    print("My name is", name[0], "and I'm", age[0], "years old")
    print("My name is", name[1], "and I'm", age[1], "years old")
    print("My name is", name[2], "and I'm", age[2], "years old")
    print("My name is", name[3], "and I'm", age[3], "years old")
    print("My name is", name[4], "and I'm", age[4], "years old")
intro("Tushar", "Himanshu", "Kartik", "Rahul", "Jatin", age=[20, 18, 11, 26, 20])
```

### for loop in fuctions:

```python
def gm(greet, *names):
    for i in names:
        print(greet, i)
gm("Good Morning", "Tushar", "Rahul", "Aman", "Priya", "Neha", "Pooja")
```

# lambda function:
### A small nameless function defined as lambda keyword.
### It can take any no. of arguments but has only one expression.

```python
a = lambda a,b: a+b
print(a(34,76))
```

```python
a = lambda a,b,c: a*b*c
print(a(34,76,89))
```

# Scopes: Area of program where a variabe can occur

## local scope:                                              
### Defines inside a function.                               
### can be used only inside a function.                      
### destroy after function execution.                        
### no keyword needed to modify.                            

```python
def myfunc(a,b):
    x = a+b        # local scope
    print(x)
myfunc(34,43)
```

```python
def myfunc():
  x = 300
  def myinnerfunc():
    print(x)
  myinnerfunc()

myfunc()
```

## Global scope:
### Defines outside all functions.
### Can be used both inside and outside of a function.
### available throughout the function.
### "global" keyword require to modify.

```python
x = 300

def myfunc():
  x = 200
  print(x)

myfunc()

print(x)
```
