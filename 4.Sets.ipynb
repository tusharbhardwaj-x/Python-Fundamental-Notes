# Sets{}:
### It is data type that store multiple values in a single variable..
### Unmutable: We can't change the values present in a set.
### Un-ordered Data: Strings and ints are diffrentiated in a output.
### Do-not holds duplicate values.
### Not accessiable.

### add: To add any argument in a set

```python
a = {1, "T", "B", "H", 78, "K", 89, 53, 4}
a.add("P")
print(a)
a.add(9500)
print(a)
```

### update: To join two sets

```python
a = {1, 2, 3, 4, 5}
b = {"T", "U", "S", "H", "A", "R"}
a.update(b)
print(a)
```

### pop: To delete any random value from the set

```python
a = {"T", "U", "S", "H", "A", "R"}
a.pop()
print(a)
a.pop()
print(a)
```

### remove: To remove any  value as you want from the given set

```python
a = {"T", "U", "S", "H", "A", "R"}
a.remove("U")
print(a)
a.remove("R")
print(a)
```

### clear: To remove all values from the set

```python
a = {"T", "U", "S", "H", "A", "R"}
a.clear()
print(a)
```

### del: To delete all values as well as structure of the given set

```python
a = {"T", "U", "S", "H", "A", "R"}
del a
print(a)
```

### loop: To ittrate all values of the set

```python
a = {"T", "U", "S", "H", "A", "R"}
for x in a:
    print(x)
```

## Joins: To join sets according to the following condition

### union(|): To  print all values of both sets

```python
a = {"T", "U", "S", "H", "A", "R"}
b = {"T", "U", 7, 2, 8, 9, 7, 8,}
c = a.union(b)
print(c)
print(a.union(b))
d = a|b
print(d)
print(a|b)
```

### intersection(&): To print only common values present in both sets

```python
a = {"T", "U", "S", "H", "A", "R"}
b = {"T", "U", 7, 2, 8, 9, 7, 8,}
c = a.intersection(b)
print(c)
print(a.intersection(b))
d = a&b
print(d)
print(a&b)
```

### difference(-): To print all values of first set which are not present in second set

```python
a = {"T", "U", "S", "H", "A", "R"}
b = {"T", "U", 7, 2, 8, 9, 7, 8,}
c = a.difference(b)
print(c)
print(a.difference(b))
d = a-b
print(d)
print(a-b)
```

### symmetric_difference(^): To print only all uncommon values present in both sets

```python
a = {"T", "U", "S", "H", "A", "R"}
b = {"T", "U", 7, 2, 8, 9, 7, 8,}
c = a.symmetric_difference(b)
print(c)
print(a.symmetric_difference(b))
d = a^b
print(d)
print(a^b)
```
