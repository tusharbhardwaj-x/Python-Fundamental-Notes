# Tuples():
### It is a datatype that stores multiple variable in a single vaariable.
### Unmutable: Values in a tuple are not changable.
### Ordered data: Give output as same as input.
### Duplicate Values are allowed.
### Tuple has only two methods(count, index)

### count: To count how many times a value repeat in a tuple

```python
a = ("T", "B", "H", 89, 73, "K", 89, "B", 89)
b = a.count(89)
print(b)
c = a.count("B")
print(c)
print(a.count("B"))
```

### index: To find indexing of given value

```python
a = ("T", "B", "H", 89, 73, "K", 89, "B", 89)
b = a.index(89)
print(b)
print(a.index("T"))
```

### To print value present on given indexing

```python
a = ("T", "B", "H", 89, 73, "K", 89, "B", 89)
b = a[4]
print(b)
print(a[7])
```

### To print the tuple with a specific range

```python
a = ("U", "Owe", "a", "car", 95000, 400, 24, "rs")
b = a[1:7:2]    # 1:start inndexing, 7:End indexing, 2:Gap b/w the outputs
print(b)
print(a[1:4])
```

### Coverting tuple into list and list into tuple

```python
a = ("U", "Owe", "a", "car", 95000, 400, 24, "rs")
b = list(a)
print(b)
c = tuple(b)
print(c)
print(list(c))
print(tuple(c))
print(tuple(list(c)))
```

### Unpacking the tuple: Use when we have more values but want less variable to ittrate.

```python
Name = ("Tushar", "Himanshu", "Kartik", "Rahul", "Deepak", "Jatin", "Rohit")
(T, H, K, *O) = Name
print(Name)
print(T)
print(H)
print(K)
print(O)  # output will be a list, you can convert it into tupple
print(tuple(O))
```
